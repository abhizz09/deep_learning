```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
from sklearn import datasets
df=datasets.load_digits()
```

```python
#
xt=df.data
yt=df.target
yt=pd.get_dummies(yt)
yt=np.array(yt)
#learning rate
LR=.001
```

```python
x,x_test,y,y_test=train_test_split(xt,yt,test_size=.30, random_state=42)
```

```python
#initialise shapes 
dim=x.shape[1]
dim
dim2=x.shape[0]
y.shape[1]
m=x.shape[0]
```

```python
#number of neurons in hidden layer 1
hidden_layer1_perceptron=5
#number of neurons in hidden layer 2
hidden_layer2_perceptron=4
#number of neurons in output layer
op_layer_perceptron=y.shape[1]
```

### describing shapes of hidden layer and out put layers


```python
w1=np.random.randn(hidden_layer1_perceptron,dim)*0.01                                                
b1=np.zeros((hidden_layer1_perceptron,1))
w2=np.random.randn(hidden_layer2_perceptron,hidden_layer1_perceptron)*0.01
b2=np.zeros((hidden_layer2_perceptron,1))
w_out=np.random.randn(op_layer_perceptron,hidden_layer2_perceptron)*0.01
b_out=np.zeros((op_layer_perceptron,1))
```

### ReLU function

```python
def ReLU(Z):
    return np.maximum(0,Z)
```

### sigmoid Function

```python
def sigmoid (x):
    return 1/(1 + np.exp(-x))
```

###  differentiation of ReLU function

```python
def diff_activation_ReLU(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
#1st layer
for i in range(100):
        #print(i)
        z=np.dot(w1,x.T)
        z=z+b1
        A1=ReLU(z)
        
        
        #layer 2

        z1=np.dot(w2,A1)
        z1=z1+b2
        A2=ReLU(z1)


        #out put layer

        z2=np.dot(w_out,A2)
        z2=z2+b_out
        Y_OUT=sigmoid(z2)

        cost=np.sum(np.multiply(y.T,np.log(Y_OUT)) + np.multiply((1-y.T),np.log(1-Y_OUT)))*(-1/m)

        #if i==8000 or i==10000 or i==12000 or i==13000 or i==14999 or i==500 or i==1000 or i==1500 or i==2000 or i==2500 or i==30000 or i==3500 or i==4000 or i==4500 or i==5000:
        print('i==',i,'cost==',cost)

        dz3=Y_OUT-y.T
        dw3=(np.matmul(dz3,A2.T))*(1/m)
        db3=np.sum(dz3,axis=1,keepdims=True)
        
       
        diff_activation_HL2=np.ones((hidden_layer2_perceptron,m))
        for i in range(hidden_layer2_perceptron):
            for j in range(m):
                 diff_activation_HL2[i,j]=diff_activation_ReLU(z1[i,j])

        dz2=np.dot(w_out.T,dz3)* diff_activation_HL2
        #print(dz2.shape)
        #print(A1.T.shape)
        dw2=(1/m)*np.dot(dz2,A1.T)
        #print(dw2.shape)
        #print("w2.T,dz2",np.dot(w2.T,dz2).shape)
        #print("z",z.shape)
        db2=np.sum(dz2,axis=1,keepdims=True)
        
        diff_activation_HL1=np.ones((hidden_layer1_perceptron,m))
        for i in range(hidden_layer1_perceptron):
            for j in range(m):
                diff_activation_HL1[i,j]=diff_activation_ReLU(z[i,j])
        dz1=np.dot(w2.T,dz2)*diff_activation_HL1
        #print("dz1",dz1.shape)
        dw1=(1/m)*np.dot(dz1,x)
        db1=np.sum(dz1,axis=1,keepdims=True)
        
        
        
        w1=w1-LR*dw1
        b1=b1-LR*db1
        w2=w2-LR*dw2
        b2=b2-LR*db2
        w_out=w_out-LR*dw3
        b_out=b_out-LR*db3
```
